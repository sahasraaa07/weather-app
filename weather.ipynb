```python
import requests

# Ask user for city name
city = input("Enter city name: ")

# Convert city name to coordinates using open-meteo geocoding API
geo_url = f"https://geocoding-api.open-meteo.com/v1/search?name={city}"
geo_response = requests.get(geo_url).json()

# Check if city exists
if "results" not in geo_response or len(geo_response["results"]) == 0:
    print("City not found. Please try another city.")
else:
    # Extract latitude and longitude
    lat = geo_response["results"][0]["latitude"]
    lon = geo_response["results"][0]["longitude"]

    print(f"\nCoordinates of {city}: {lat}, {lon}")

    # Fetch weather data
    weather_url = (
        f"https://api.open-meteo.com/v1/forecast?"
        f"latitude={lat}&longitude={lon}&current_weather=true"
    )

    weather_response = requests.get(weather_url).json()

    # Safe check
    if "current_weather" in weather_response:
        current = weather_response["current_weather"]

        print("\n--- Current Weather ---")
        print("Temperature:", current.get("temperature", "N/A"), "°C")
        print("Wind Speed:", current.get("windspeed", "N/A"), "km/h")
        print("Weather Code:", current.get("weathercode", "N/A"))
        print("------------------------")
    else:
        print("Weather data not available for this city.")
```

# Weather App in Python

This project fetches real-time weather data for any city using the free Open-Meteo API.

### Features:
- Convert city name to coordinates
- Get current temperature
- Get wind speed
- No API key required
